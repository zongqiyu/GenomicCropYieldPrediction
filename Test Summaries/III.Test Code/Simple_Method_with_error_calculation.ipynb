```python
# matrix completion!!! Mapping the gene lines to new environments
import numpy as np
import csv

def swap_columns(matrix, col1, col2):
    for row in matrix:
        row[col1], row[col2] = row[col2], row[col1]

mat_name2 = 'RPunch1415_Original.csv'
with open(mat_name2) as f:
  reader2 = csv.reader(f)
  lst2 = list(reader2)
  original = np.array(lst2)
  #print(mat)

original = np.delete(original,0,0)
original = np.delete(original,0,1)
original = original.astype(float)
mat = original.copy()
mat[0:194,4] = 0
swap_columns(mat,4,4)
swap_columns(original,4,4)
#print(mat)
#print(original)
```

```python
hori_piece = mat[195:775,:]
verti_piece = mat[:,0:4]
key_piece = mat[195:775,0:4]

inv = np.linalg.pinv(key_piece)
#print(inv)

SimpleMap = verti_piece @ inv @ hori_piece
#print(hori_piece.shape)
#print(verti_piece.shape)
#print(key_piece.shape)
```

```python
sqEr = 0

for y in range(mat.shape[1]): # change according to size
  for x in range(mat.shape[0]): # change according to size
    if(mat[x,y]== 0):
      sqEr = sqEr + abs(1 - SimpleMap[x,y]/original[x,y])
      simple_error_doc.append(1 - SimpleMap[x,y]/original[x,y])

print(sqEr)
```

```python
# Simple Method for completing blocked matrix
dim = mat.shape
count_hori = 0
count_verti = 0

for i in range(dim[1]): # change according to horizontal size
  if(mat[0,i] != 0):
    print(i)
    count_hori = i
print(mat[0,3])

for j in range(dim[0]): # change according to vertical size
  if(mat[j,dim[1]-1] != 0):
    count_verti = count_verti + 1
print(count_verti)

hori_piece = np.zeros([count_verti,dim[1]])  # change numbers of columns
verti_piece = np.zeros([dim[0],count_hori]) # change numbers of rows
key_piece = np.zeros([count_verti, count_hori])
#print(hori_piece)
#print(verti_piece)
#print(key_piece)

for i in range(dim[0] - count_verti,dim[0]):
  for j in range(dim[1]):
    hori_piece[i - dim[0] + count_verti,j] = mat[i,j]

for i in range(dim[0]):
  for j in range(0, count_hori):
    verti_piece[i,j] = mat[i,j]

for i in range(dim[0] - count_verti,dim[0]):
  for j in range(count_hori):
    key_piece[i - dim[0] + count_verti,j] = mat[i,j]

#print(hori_piece)
#print(verti_piece)
#print(key_piece)

inv = np.linalg.pinv(key_piece)
#print(inv)

SimpleMap = verti_piece @ inv @ hori_piece
#print(hori_piece.shape)
```
